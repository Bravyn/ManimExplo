The code below is complicated for a beginner, for now just try scheming it and noticing some of the python constructs used. We have classes going on and  stuff. The first line In Colab, the `%%manim` magic command allows you to run Manim directly in a code cell. Let's break down the meaning of the flags used in the command `%%manim -qm -v WARNING SquareToCircle`:

- `%%manim`: This magic command indicates that the following code cell should be executed using Manim.

- `-qm`: This flag stands for "quality medium" and sets the rendering quality to a medium level. It strikes a balance between good visual quality and reasonable rendering time.

- `-v WARNING`: This flag sets the verbosity level to "WARNING", which means that only important warnings will be displayed during the rendering process. It reduces the amount of output shown in the notebook.

- `SquareToCircle`: This is the name of the class or function in the code cell that contains the animation to be rendered. In this case, it suggests that there may be an animation transforming a square into a circle.

So, when you execute the `%%manim -qm -v WARNING SquareToCircle` command in the code cell, Manim will render the animation defined in the `SquareToCircle` class, using medium rendering quality and displaying only important warnings during the process.

```python
%%manim -qm -v WARNING SquareToCircle


class SquareToCircle(Scene):
   def construct(self):
      square = Square()#creates a square with a default length l, pass in some l
      circle = Circle()#creates a circle with a default radius r, pass in some r
      circle.set_fill(YELLOW, opacity=0.7)
      self.play(Create(square))
      self.play(Create(circle))
      self.play(Transform(circle, square))#transform is some sort of morphing 
      self.wait()#waits for a given time, default is 1s
```

Can you spot the difference between the code and the one below? The clue is in the animation produced!

```python

%%manim -qm -v WARNING Addition

class Addition(Scene):
   def construct(self):
      square = Square()
      circle = Circle()
      circle.set_fill(YELLOW, opacity=0.7)
      self.play(Create(circle))
      self.play(Transform(circle, square))
      self.wait()#be creative with this function, timing is everything!!!
```

More code to explore, play with the opacity and add a wait() function to add pauses and suspense!

```python
%%manim -qm -v WARNING Smile

class Smile(Scene):
  def construct(self): 
    circle = Circle()
    square = Square()

    self.play(Create(circle))
    self.play(Create(square))
    circle.set_fill(BLUE, opacity = .9)
    self.play(Create(circle))
    self.wait(2)
    square.set_fill(WHITE, opacity = .3)
    self.play(Transform(circle, square))
    self.wait(3)
```

The manim library has a way of doing things, fundamental to them is the idea of creating animations with code, specifically with the construct method. It is where all the animation code goes.

# Bringing it all together

Now let's combine some text and some shapes to see what happens.

```python
%%manim -qm -v WARNING ShowAll

class ShowAll(Scene):
  def construct(self):
    text = Text("Let us combine our knowledge so far...", font_size= 36)
    text2 = Text("Now to the circle...", font_size= 36)
    circle = Circle(2);
    square = Square(4)#4 is the side length of our square
    self.add(text)
    self.wait(2)
    self.remove(text)
    self.play(Create(square))
    self.wait(3)
    self.add(text2)
    self.wait(3)
    self.remove(text2)
    #self.wait(1)
    self.play(Transform(text2, circle))
    self.wait(5)

                               
```

Now that works!But the text could be better of on the side. So let us do that.

```python
%%manim -qm -v WARNING NormalText

class NormalText(Scene):
  def construct(self):
    self.add(Text("Hello, I am Mr. Normal Text."))
```

```python
%%manim -qm -v WARNING TextAside

class TextAside(Scene):
  def construct(self):
    text1 = Text("Hello, I am Ms. Gradient Color Text.", \
                  gradient = (RED, YELLOW, BLUE))
    self.add(text1)
    self.add(Text("Ain't I beautiful?", \
                  gradient = (RED, YELLOW, BLUE), font_size = 35).next_to(text1, DOWN))
    self.play(Create(Circle(radius = 8, color =  YELLOW).next_to(text1, LEFT)))
    self.play(Create(Square(side_length = 1.5, color =  YELLOW).next_to(text1, LEFT)))
    self.wait(2)
    self.play(Create(Square(side_length = 2.5, color =  WHITE).next_to(text1, RIGHT)))
```

```python
%%manim -qm -v WARNING Status 

class Status(Scene):
  def construct(self):
    text1 = Text("Hi")
    square = Square(4.0, color = YELLOW_A)
    self.add(text1)
    self.play(Create(square))
    self.wait()
    self.remove(text1)
    self.remove(square)
    self.wait()
    text2 = Text("Welcome")
    circle1 = Circle(3.0, color = YELLOW_B)
    self.add(text2)
    self.play(Create(circle1))
    text3 = Text("My Name is Ian")
    self.remove(text2)
    self.wait()
    self.add(text3)
    self.wait(2)
    self.remove(text3)
    self.remove(circle1)
    topic = Text("Today, we are going...")
    self.add(topic)
    self.wait(2)
    self.remove(topic)
    topic2 = Text("To Learn About Rust", line_spacing = 2, color = YELLOW)
    self.add(topic2)
    self.wait(3)

```
